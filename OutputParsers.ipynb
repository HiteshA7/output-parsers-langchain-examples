````python
from typing import TypedDict,Literal,Annotated,Optional
from pydantic import BaseModel,Field
````

````python
class DataFormat(BaseModel):
  name : str = Field(description="you have to generate the name of the person",default="raj")
  age : Optional[int] = "jasjd","dwd"
  summary : Annotated[str,Field(description = "you have to generate the summary of the person",default="dswd")]
````

````python
DataFormat()
````

````python
jsop = JsonOutputParser(pydantic_object=DataFormat)
````

````python
jsop.get_format_instructions()
````

````python
ca = model.invoke([{"role":"system","content" :"""You are a 10 year experienced model and now you have to as human and json format is ''' STRICT OUTPUT FORMAT:
- Return only the JSON value that conforms to the schema. Do not include any additional text, explanations, headings, or separators.
- Do not wrap the JSON in Markdown or code fences (no ``` or ```json).
- Do not prepend or append any text (e.g., do not write "Here is the JSON:").
- The response must be a single top-level JSON value exactly as required by the schema (object/array/etc.), with no trailing commas or comments.

The output should be formatted as a JSON instance that conforms to the JSON schema below.

As an example, for the schema {"properties": {"foo": {"title": "Foo", "description": "a list of strings", "type": "array", "items": {"type": "string"}}}, "required": ["foo"]} the object {"foo": ["bar", "baz"]} is a well-formatted instance of the schema. The object {"properties": {"foo": ["bar", "baz"]}} is not well-formatted.

Here is the output schema (shown in a code block for readability only — do not include any backticks or Markdown in your output):
```
{"properties": {"name": {"default": "raj", "description": "you have to generate the name of the person", "title": "Name", "type": "string"}, "age": {"anyOf": [{"type": "integer"}, {"type": "null"}], "default": ["jasjd", "dwd"], "title": "Age"}, "summary": {"default": "dswd", "description": "you have to generate the summary of the person", "title": "Summary", "type": "string"}}}
```"""},
              {"role":"human","content":"generate the required things and also the summary must be about tom and jerry "}])
````

````python
jop.parse(ca.content)
````
